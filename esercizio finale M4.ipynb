```python
import pandas as pd
covid=pd.read_csv("owid-covid-data.csv",sep=",")
print(covid)
```

```python

print(covid.shape)

print(covid.columns)

print(covid.describe())
```

Per ogni continente trovare il numero di casi totali avvenuti in quello stesso continente; non considerare locazioni che nel dataset non appartengono a nessun continente.

```python
continenti=covid.groupby(["continent","location"])["total_cases"].aggregate("max").groupby("continent").aggregate("sum")
print(continenti)
```

Sempre riguardo ai casi di Covid totali, sviluppare una funzione che prenda in input il dataset e due nomi di continenti e che ne confronti i seguenti relativi descrittori statistici:valori minimo e massimo, media e percentuale rispetto al numero dei casi totali nel mondo.

```python
def inscont():
  continente=str(input("Inserire il nome di un continente a scelta fra: Africa, Asia, Europe, Oceania, North Amarica, South America: "))
  mask_continente=covid["location"]==continente
  df_continente=covid[mask_continente]
  minimo=df_continente["total_cases"].min()
  massimo=df_continente["total_cases"].max()
  media=df_continente["total_cases"].mean()
  mask_mondo=covid["location"]=="World"
  df_mondo=covid[mask_mondo]
  percentuale=100*massimo/df_mondo["total_cases"].max()
  print("Il minimo del primo continente è:",minimo)
  print("Il massimo del primo continente è:", massimo)
  print("La media del primo continente è:", media)
  print("la percentuale del primo continente è:", percentuale)
  continente=str(input("Inserire il nome di un continente a scelta fra: Africa, Asia, Europe, Oceania, North Amarica, South America: "))
  mask_continente=covid["location"]==continente
  df_continente=covid[mask_continente]
  minimo=df_continente["total_cases"].min()
  massimo=df_continente["total_cases"].max()
  media=df_continente["total_cases"].mean()
  percentuale=100*massimo/df_mondo["total_cases"].max()
  print("Il minimo del secondo continente è:",minimo)
  print("Il massimo del secondo continente è:", massimo)
  print("La media del secondo continente è:",media)
  print("La percentuale del secondo continente è:",percentuale)
inscont()


```

```python
def insvaccinations():
  continente=str(input("Inserire il nome di un continente a scelta fra: Africa, Asia, Europe, Oceania, North Amarica, South America "))
  mask_continente=covid["location"]==continente
  df_continente=covid[mask_continente]
  minimo=df_continente["total_vaccinations"].min()
  massimo=df_continente["total_vaccinations"].max()
  media=df_continente["total_vaccinations"].mean()
  mask_mondo=covid["location"]=="World"
  df_mondo=covid[mask_mondo]
  percentuale=100*massimo/df_mondo["total_vaccinations"].max()
  print("Il minimo del primo continente è:",minimo)
  print("Il massimo del primo continente è:", massimo)
  print("La media del primo continente è:", media)
  print("la percentuale del primo continente è:", percentuale)
  continente=str(input("Inserire il nome di un continente a scelta fra: Africa, Asia, Europe, Oceania, North Amarica, South America "))
  mask_continente=covid["location"]==continente
  df_continente=covid[mask_continente]
  minimo=df_continente["total_vaccinations"].min()
  massimo=df_continente["total_vaccinations"].max()
  media=df_continente["total_vaccinations"].mean()
  percentuale=100*massimo/df_mondo["total_vaccinations"].max()
  print("Il minimo del secondo continente è:",minimo)
  print("Il massimo del secondo continente è:", massimo)
  print("La media del secondo continente è:",media)
  print("La percentuale del secondo continente è:",percentuale)
insvaccinations()
```

Dall'analisi dei dati effettuata emerge che tra i continenti di cui viene richiesta l'analisi - Europa, Oceania e Sud America- l'Europa è quello che presenta il maggior numero di casi totali di covid rispetto al numero della propria popolazione, nel raffronto con Oceania e Sud America. Nel contempo, però, presenta anche il maggior numero complessivo di vaccinazioni. Al contrario l'Oceania presenta la minore percentuale di soggetti vacccinati.
